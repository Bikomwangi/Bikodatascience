```
# read the telcom data sets as data frames using pandas
```

```
# telco data set 1
url = 'http://bit.ly/Telcom_dataset1'
telco_dataset1_df = pd.read_csv('http://bit.ly/Telcom_dataset1')
telco_dataset1_df
```

```
# telco dataset 2
url = 'http://bit.ly/Telcom_dataset2'
telco_dataset2_df = pd.read_csv('http://bit.ly/Telcom_dataset2')
telco_dataset2_df
```

```
# telco dataset 3
url = 'http://bit.ly/Telcom_dataset3'
telco_dataset3_df = pd.read_csv('http://bit.ly/Telcom_dataset3')
telco_dataset3_df
```

```
# deleting some columns from cells_geo
cells_geo_df.drop(['DECOUPZONE', 'LONGITUDE', 'LATITUDE', 'AREA'], axis=1, inplace=True)
cells_geo_df

```

```
# renaming some variables in telco dataset 3
telco_dataset3_df.rename(columns={'CELLID': 'CELL_ID', 'SIET_ID': 'SITE_ID'}, inplace=True)
telco_dataset3_df
```

```
# renaming DW A NUMBER column in dataset 2 to match with the rest of the dataframes
telco_dataset2_df.rename(columns={'DW_A_NUMBER': 'DW_A_NUMBER_INT', 'DW_B_NUMBER': 'DW_B_NUMBER_INT'}, inplace=True)
telco_dataset2_df
```

```

#rename PRODUTC in telco data set 1
telco_dataset1_df.rename(columns={'PRODUTC':'PRODUCT'}, inplace=True)
telco_dataset1_df
```

```
# merging telcom data sets
telcom_dataset_df = pd.concat((telco_dataset1_df, telco_dataset2_df, telco_dataset3_df), axis=0, ignore_index=True)
telcom_dataset_df
```

```
# finding out the most cities used for the three days
# and drop null values
cells_geo_df.dropna(inplace = True)
most_cities_used = cells_geo_df['VILLES'].value_counts()#.idmax()
most_cities_used
```
